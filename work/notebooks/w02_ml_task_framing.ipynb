## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule system (like standard IF/ELSE statements) cannot handle the complexity of this problem. The target score is influenced by dozens of interacting, non-linear features that change dynamically over time. Hardcoding thresholds (e.g. if feature_A > 50 and feature_B == True) would be brittle, impossible to maintain, and would completely fail to accurately capture subtle differences between a score of 75 vs. 80. Machine learning allows us to learn these complex patterns automatically from data.
